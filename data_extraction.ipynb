# Data Extraction

```python
import requests
```

We extract the data using the API, focusing only on the years we are interested in.

```python
url = 'https://esploradati.istat.it/SDMXWS/rest/data/41_983'
header = {'Accept': 'application/vnd.sdmx.data+csv;version=1.0.0'}
params = {'startPeriod':'2020',
          'endPeriod':'2024'}

r = requests.get(url,headers=header,params=params)
r.status_code
```

Now we save the .csv file into the data folder.

```python
with open('data/crashes.csv','w') as f:
    f.write(r.text)
```
